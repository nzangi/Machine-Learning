```python
#importing depedencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#load the dataset into pandas
credit_data = pd.read_csv("///home/nzangi/Downloads/youtube-dl/ML projects/Datasets/creditcard.csv")
```

```python
#checking the missing values
credit_data.isnull().sum()
```

```python
#check the distribtion of legit transaction and fraud
credit_data['Class'].value_counts()
```

```python
#the data set is Highly Unbalanced
#0---> Normal transaction
#1---> fraud transaction
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]


```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#get statical measures of the data of legit data
legit.Amount.describe()
```

```python
#get all the satistical data of the fraud trasmission
fraud.Amount.describe()
```

```python
#compare the values of the both transimission
credit_data.groupby('Class').mean()
```

```python
#Building a dataset which will contain the equal distribution both normal fraud transmission data
legit_sample = legit.sample(n=492)
```

```python
#contacting the two data set
#axis = 0 means add values in rows
#axis = 1 means add values in coumns
new_dataset = pd.concat([legit_sample,fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns="Class", axis = 1)
Y = new_dataset["Class"]
```

```python
print(X)
print(Y)
```

```python
#split the data int train and the test data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, stratify = Y,random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#creating a model
model = LogisticRegression()
```

```python
#training our model
model.fit(X_train,Y_train)
```

```python
#Evaluation of the model
#accuracy on the trainng data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(f"The training data accuracy is:{training_data_accuracy} ")
```

```python
#accuracy on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(f"The test data accuracy is:{test_data_accuracy} ")
```
